La tarea es obtener un modelo de lenguaje basado en bigramas que nos permita calcular la probabilidad de una palabra dada la anterior.

```
from sentence_splitter import SentenceSplitter
splitter = SentenceSplitter(language = 'es')

oraciones = splitter.split(texto)
```

```
oraciones_tokenizadas = []
for oracion in oraciones:
   tokens = nltk.word_tokenize(oracion,"spanish")
   oraciones_tokenizadas.append(tokens)
```

```
oraciones_tokenizadas[0][0]
```

```
for i, oracion in enumerate(oraciones_tokenizadas):
  oraciones_tokenizadas [i] = ["<s>"] + oracion + ["</s>"]
```

```
print(oraciones_tokenizadas[0])
```

La lista de oraciones

```
bigramas_oraciones = []
unigramas_oraciones = []

for oracion in oraciones_tokenizadas:
  bigramas = nltk.ngrams(oracion,2)
  bigramas_oraciones.append(bigramas)
  unigramas = nltk.ngrams(oracion,1)
  unigramas_oraciones.append(unigramas)
```

```
for gramas in unigramas_oraciones [0]:
  print(gramas)
```

```
unigramas = [j for i in unigramas_oraciones for j in i]
bigramas = [ j for i  in bigramas_oraciones for j in i]
```

para hacer los conteos rápidos (frecuencias) utilizamos coleections

```
import collections

frecuencia_unigramas = collections.Counter(unigramas)
frecuencia_bigramas = collections.Counter(bigramas)
frecuencia_bigramas
frecuencia_unigramas
```

```
numerador = frecuencia_bigramas[('hacerlo','callar')]
denominador = frecuencia_unigramas[('hacerlo',)]

print(numerador/denominador)
```

para probabilidades condicionales (probabilidad de x dado y utilizamos Conditional FreqDist


```
from nltk.probability import ConditionalFreqDist
frecuencias_condicionales = ConditionalFreqDist()
condicion = ('hacerlo',)

for unigrama in frecuencia_unigramas.keys():
  frecuencia = frecuencia_bigramas[condicion + unigrama]
  if frecuencia:
    frecuencias_condicionales[condicion][unigrama] = frecuencia
```

```
palabra_mayor =frecuencias_condicionales[condicion].max()
print(palabra_mayor)
```
